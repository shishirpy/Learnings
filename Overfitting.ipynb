```python
"""
Created on Sat Mar 30 17:44:49 2013

@author: shishir
@email: shishir.py@gmail.com

This program demostrates the concept of overfitting. It is based on the
material in the book - "Learning from Data" by Mostafa, Ismail and Lin.
The authors have done a great job of explaning the concept of overfitting.
You can view their online lecture on overfitting at - 
http://www.youtube.com/watch?v=EQWr3GGCdzw&hd=1#t=3m43s


Things to try out:
See what happens to the out of sample error when the number of samples 
is increased. 
"""
print __doc__

import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.legendre as Leg
import pdb

N = 15      #number of samples
d = 50      #degree of the targe polynomial
M  = 1000   #linspace number of spacings
sigma = 0.5

r_min, r_max = -0.5,0.5


cs = np.random.uniform(-1,1,d+1)  #Legendre polynomial coefs

x = np.random.uniform(low=r_min, high = r_max, size = N)
y = Leg.legval(x,cs)
t = sigma * np.random.randn(N)
y = y + t

#correct polynomial
x1 = np.linspace(r_min,r_max, M)
y1 = Leg.legval(x1,cs)


#2nd degree polynomial fit
z2 = np.polyfit(x,y,2)
y2 = np.polyval(z2,x1)

#10 degree polynomial fit
z10 = np.polyfit(x,y,10)
y10 = np.polyval(z10, x1)

#calculate errors
Ein2 = np.average(np.linalg.norm(y - np.polyval(z2, x))**2)
Ein10 = np.average(np.linalg.norm(y - np.polyval(z10, x))**2)

Eout2 = np.average(np.linalg.norm(y1 - np.polyval(z2,x1))**2)
Eout10 = np.average(np.linalg.norm(y1 - np.polyval(z10,x1))**2)

print "-" * 62
print " " * 20 + "| Degree 2 fit" + " "*6 + "|" + " " * 6 + "Degree 10 fit |"
print "-" * 62
print "In sample Error" + " "* 5+ "|", Ein2, " " * 4 + "|" + " "*  4 , Ein10, "|"
print "-" * 62
print "Out of sample Error |", Eout2, " " * 4 + "|" + " "*  4 , Eout10, "|"
print "-" * 62

print "\n"
print "Overfitting value", Eout10 - Eout2


fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.plot(x,y, 'ko', x1, y1, 'b-', x1, y2, 'g--')#, x1, y10, 'r-')
ax.plot(x1,y10, 'r-')
ax.grid('on')
ax.set_title("Polynomial generated by Legendre polynomials")
ax.set_ylim(-5, 5)
ax.legend(('Data Points', 'True Polynomial', '2 Degree fit', '10 degree fit'),
          'upper center', shadow = 'True')
plt.show()
```
