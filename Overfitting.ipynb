# Overfitting in Machine Learning

In this notebook I demostrate the concept of overfitting based on the material from the book - "Learning from Data" by Mostafa, Ismail and Lin. The will try a polynomial fit on our data. The data will be generated from a random polynomial that is of degree 50.

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.legendre as Leg
```

### Notation

$N \quad$ represents the number of data points.

$d \quad$ degree of the target polynomial (true polynomial to generate the data points).

$\sigma \quad$ std. deviation of the Gaussian noise.

We are assuming the following linear model 
$$y = X \beta + \eta $$

Where:

$\eta \quad$ is the Gaussian noise


```python
N = 15     #number of samples
d = 50      #degree of the targe polynomial
M  = 1000   #linspace number of spacings
sigma = 0.5

r_min, r_max = -1,1  #range from which the points will be picked
```

```python
def Polyfit(N, d, sigma, M):
    np.random.seed(5)
    cs = np.random.uniform(-1,1,d+1)  #True Legendre polynomial coefs
    
    x = np.random.uniform(low=r_min, high = r_max, size = N)
    y = Leg.legval(x,cs)
    t = sigma * np.random.randn(N)
    y = y + t
    
    #correct polynomial
    x1 = np.linspace(r_min,r_max, M)
    y1 = Leg.legval(x1,cs)
    
    
    #2nd degree polynomial fit
    z2 = np.polyfit(x,y,2)
    y2 = np.polyval(z2,x1)
    
    #10 degree polynomial fit
    z10 = np.polyfit(x,y,10)
    y10 = np.polyval(z10, x1)
    
    #calculate errors
    Ein2 = np.average(np.linalg.norm(y - np.polyval(z2, x))**2)
    Ein10 = np.average(np.linalg.norm(y - np.polyval(z10, x))**2)
    
    Eout2 = np.average(np.linalg.norm(y1 - np.polyval(z2,x1))**2)
    Eout10 = np.average(np.linalg.norm(y1 - np.polyval(z10,x1))**2)
    
    print "-" * 62
    print " " * 20 + "| Degree 2 fit" + " "*6 + "|" + " " * 6 + "Degree 10 fit |"
    print "-" * 62
    print "In sample Error" + " "* 5+ "|", Ein2, " " * 4 + "|" + " "*  4 , Ein10, "|"
    print "-" * 62
    print "Out of sample Error |", Eout2, " " * 4 + "|" + " "*  4 , Eout10, "|"
    print "-" * 62
    
    print "\n"
    print "Overfitting value", Eout10 - Eout2
    
    
    matplotlib.rcParams['figure.figsize'] = (15,10)
    
    fig = plt.figure(1)
    ax = fig.add_subplot(111)
    ax.plot(x,y, 'kx', x1, y1, 'b-', x1, y2, 'g--')#, x1, y10, 'r-')
    ax.plot(x1,y10, 'r-')
    ax.grid('on')
    ax.set_title("Polynomial generated by Legendre polynomials")
    ax.set_ylim(-5, 5)
    ax.legend(('Data Points', 'True Polynomial', '2 Degree fit', '10 degree fit'),
              'upper center', shadow = 'True')
    plt.show()
```

```python
Polyfit(N, d, sigma, M)
```

```python
Polyfit(50, d, sigma, M)
```

```python
Polyfit(5000000, d, sigma, M)
```
